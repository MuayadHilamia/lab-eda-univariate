```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
df.head()
```

```python
# 1.1
categories_count = df["category"].value_counts()
top_categories = categories_count.head(5)
```

```python
# 1.2 
plt.figure(figsize=(10,5))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title("Top 5 Product Categories on Amazon UK")
plt.ylabel("Number of Listings")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2.
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

var_price = df['price'].var()
std_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance: {var_price:.2f}")
print(f"Standard Deviation: {std_price:.2f}")
print(f"Range: {range_price:.2f}")
print(f"IQR: {iqr_price:.2f}")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[df['price'] < 500]['price'], bins=40, kde=True)
plt.title("Price Distribution (Filtered Under 500)")
plt.xlabel("Price ()")
plt.show()
```

```python
# -------- 1. Measures of Centrality --------

mean_stars = df['stars'].mean()      
median_stars = df['stars'].median()  
mode_stars = df['stars'].mode()[0]   

print(f"Mean: {mean_stars:.2f}, Median: {median_stars:.2f}, Mode: {mode_stars:.2f}")
```

```python
var_stars = df['stars'].var()
std_stars = df['stars'].std()
iqr_stars = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print(f"Variance: {var_stars:.3f}")
print(f"Standard Deviation: {std_stars:.3f}")
print(f"IQR: {iqr_stars:.3f}")
```

```python
skew_stars = df['stars'].skew()
kurt_stars = df['stars'].kurt()

print(f"Skewness: {skew_stars:.3f}")
print(f"Kurtosis: {kurt_stars:.3f}")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['stars'], bins=20, kde=True, color="gold")
plt.title("Distribution of Product Stars (Ratings)")
plt.xlabel("Stars")
plt.ylabel("Frequency")
plt.show()
```
